```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/total-ghg-emissions.csv")
```

```python
df.head()
```

```python
fil_df= df[df['Year']== 2010]
```

```python
fil_df.head()
```

```python
df=pd.read_csv("/content/population.csv")
fil_df2= df[df['Year']== 2010]
fil_df2.head()
```

```python
df=pd.read_csv("/content/gdp-per-capita-worldbank.csv")
fil_df3= df[df['Year']== 2010]
fil_df3.head()
```

```python
square_marker_countries = [
    "United Arab Emirates", "Kuwait", "Australia", "Bahrain", "United States",
    "Canada", "Luxembourg", "New Zealand", "Ireland", "Estonia", "Saudi Arabia",
    "Singapore", "Belgium", "Netherlands", "Finland", "Denmark", "Norway",
    "Germany", "Israel", "United Kingdom", "Japan", "Slovenia", "Spain",
    "France", "Sweden", "Switzerland", "Hong Kong"
]
circle_marker_countries = [
    "Trinidad", "Turkmenistan", "Czech Republic", "Russia", "Oman", "Kazakhstan",
    "Greece", "South Korea", "Cyprus", "Austria", "Iceland", "Venezuela",
    "Poland", "Slovakia", "Ukraine", "South Africa", "Belarus", "Argentina",
    "Uruguay", "Portugal", "Hungary", "Bulgaria", "Croatia", "Latvia",
    "Lithuania", "Malaysia", "Turkey", "Chile", "Romania", "Brazil", "China",
    "Mexico", "Panama", "Costa Rica", "Bosnia", "Macedonia", "Albania", "India"
]
df_filtered_squares = fil_df[fil_df['Entity'].isin(square_marker_countries)]
df_filtered_circles = fil_df[fil_df['Entity'].isin(circle_marker_countries)]
```

```python
df_filtered_squares2 = fil_df2[fil_df2['Entity'].isin(square_marker_countries)]
df_filtered_circles2 = fil_df2[fil_df2['Entity'].isin(circle_marker_countries)]
```

```python
df_filtered_squares3 = fil_df3[fil_df3['Entity'].isin(square_marker_countries)]
df_filtered_circles3 = fil_df3[fil_df3['Entity'].isin(circle_marker_countries)]
```

```python
merged_df = pd.merge(df_filtered_squares, df_filtered_squares2 , on='Entity')
merged_df1 = pd.merge(merged_df, df_filtered_squares3, on='Entity')
merged_df1.head()
```

```python
merged_df = pd.merge(df_filtered_circles, df_filtered_circles2 , on='Entity')
merged_df2 = pd.merge(merged_df, df_filtered_circles3, on='Entity')
merged_df2.head()
```

```python
merged_df2=merged_df2[['Entity','Annual greenhouse gas emissions in CO₂ equivalents','GDP per capita, PPP (constant 2017 international $)','Population (historical)']]
```

```python
merged_df2.head()
```

```python
merged_df1=merged_df1[['Entity','Annual greenhouse gas emissions in CO₂ equivalents','GDP per capita, PPP (constant 2017 international $)','Population (historical)']]
merged_df1.head()
```

```python

merged_df1['Annual greenhouse gas emissions in CO₂ equivalents/per person'] = merged_df1['Annual greenhouse gas emissions in CO₂ equivalents'] / merged_df1['Population (historical)']

# Drop the columns 'a' and 'b'
merged_df1 = merged_df1.drop(columns=['Annual greenhouse gas emissions in CO₂ equivalents', 'Population (historical)'])
```

```python

merged_df1.head()
```

```python
merged_df2['Annual greenhouse gas emissions in CO₂ equivalents/per person'] = merged_df2['Annual greenhouse gas emissions in CO₂ equivalents'] / merged_df2['Population (historical)']

# Drop the columns 'a' and 'b'
merged_df2 = merged_df2.drop(columns=['Annual greenhouse gas emissions in CO₂ equivalents', 'Population (historical)'])
merged_df2.head()
```

```python
df_filtered_squares=merged_df1
df_filtered_circles=merged_df2
```

```python
df_filtered_circles.head()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df_filtered_squares['GDP per capita, PPP (constant 2017 international $)'], df_filtered_squares['Annual greenhouse gas emissions in CO₂ equivalents/per person'],
            color='blue', marker='s', s=50, label='Square Markers')

plt.scatter(df_filtered_circles['GDP per capita, PPP (constant 2017 international $)'], df_filtered_circles['Annual greenhouse gas emissions in CO₂ equivalents/per person'],
            color='green', marker='o', s=50, label='Circle Markers')

for i, row in df_filtered_squares.iterrows():
    plt.text(row['GDP per capita, PPP (constant 2017 international $)'], row['Annual greenhouse gas emissions in CO₂ equivalents/per person'], row['Entity'], fontsize=8, ha='right')

for i, row in df_filtered_circles.iterrows():
    plt.text(row['GDP per capita, PPP (constant 2017 international $)'], row['Annual greenhouse gas emissions in CO₂ equivalents/per person'], row['Entity'], fontsize=8, ha='right')

plt.xlabel('GDP per capita')
plt.ylabel('Annual greenhouse gas emissions in CO₂ equivalents/per person')
plt.title('Annual Greenhouse Gas Emissions vs GDP per Capita with Different Markers')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df_sum_alt = merged_df1.add(merged_df2)
output_csv = 'co2GDP_2010.csv'
df_sum_alt.to_csv(output_csv, index=False)
```
